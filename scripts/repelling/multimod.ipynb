```python
import matplotlib.pyplot as plt
import numpy as np
import ioh 

from repelling import get_meshgrid, rastrigin_function, ModifiedHimmelblau
```

```python
def rastrigin_function(x: np.ndarray) -> float:
    x = np.asarray(x)
    return 10 * len(x) + np.sum(x * x - (10 * np.cos(2 * np.pi * x)))

X, Y, Z = get_meshgrid(rastrigin_function, -1, 2)
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(projection='3d')

ax.contour(X, Y, Z, zdir='z', offset=0, cmap='coolwarm')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import qmc
```

```python
n = 64 
np.random.seed(n)
U = np.random.uniform(0, 1, size=(n, 2))
sampler = qmc.Sobol(2, scramble=False)
S = sampler.random(n)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.scatter(U[:, 0], U[:, 1])
ax2.scatter(S[:, 0], S[:, 1])

for ax in (ax1, ax2):
    ax.grid()
    ax.set_xlabel(r"$x_1$", fontsize=15)
    ax.set_ylabel(r"$x_2$", fontsize=15)

ax1.set_title(R"Random")
ax2.set_title(R"Sobol")
plt.tight_layout()
```
